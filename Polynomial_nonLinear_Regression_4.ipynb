```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

#y = w0 + w1x + w2x^2 + w3x^3 + w4x^4

#Importing the dataset
##x = data, y = quadratic equation
x_1 = np.array(3 * np.random.rand(100, 1) - 2) 
x1 = x_1.reshape(-1, 1) 
y_1 =  x_1 ** 4 +  x_1 ** 3 +  x_1 **2 +  x_1 + 11 

x_2 = np.array(5 * np.random.rand(100, 1) - 3)
x2 = x_2.reshape(-1, 1)
y_2 =  x_2 ** 4 +  x_2 ** 3 +  x_2 **2 +  x_2 + 3

#data points

plt.scatter(x_1, y_1, s = 12, color = 'y')  
plt.scatter(x_2, y_2, s = 8, color = 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Data') 
```

```python
#Model Initialization
regression_model = LinearRegression()
#Fit the data(train th model)
regression_model.fit(x1, y_1)
regression_model.fit(x2, y_2)
print('Slope of the line is', regression_model.coef_)
print('Intercet value is', regression_model.intercept_)
# Predict 
y_predicted_1 = regression_model.predict(x1) 
y_predicted_2 = regression_model.predict(x2)
```

```python
#data points
plt.scatter(x_1, y_1, s = 10)
plt.scatter(x_2, y_2, s = 8)
plt.xlabel('$x$', fontsize = 18)
plt.ylabel('$y$', rotation = 0, fontsize = 18)
plt.title('data points')

plt.plot(x_1, y_predicted_1, color = 'k')
plt.plot(x_2, y_predicted_2, color = 'y')
```

```python
#model evaluation
mse_1 = mean_squared_error(y_1, y_predicted_1)
mse_2 = mean_squared_error(y_2, y_predicted_2)
rmse_1 = np.sqrt(mean_squared_error(y_1, y_predicted_1))
rmse_2 = np.sqrt(mean_squared_error(y_2, y_predicted_2))
r2 = r2_score(y_1, y_predicted_1)
r2 = r2_score(y_2, y_predicted_2)
print('MSE of Linear model', mse_1)
print('MSE of Linear model', mse_2)
print('R2 score of Linear', r2)
print('R2 score of Linear', r2)
```

```python
poly_features = PolynomialFeatures(degree = 4, include_bias = False)
x_1_poly = poly_features.fit_transform(x1)
x_1[3]
x_2_poly = poly_features.fit_transform(x2)
x_2[3]
print(x_1[3])
print(x_2[3])
```

```python
x_1_poly[3]
x_2_poly[3]
print(x_1_poly[3])
print(x_2_poly[3])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_1_poly, y_1)
lin_reg.fit(x_2_poly, y_2)
print('Coefficients of x_1 are', lin_reg.coef_)
print('Coefficients of x_2 are', lin_reg.coef_)
print('Intercept is', lin_reg.intercept_)
```

```python
x_1_new = np.linspace(-2, 2, 100).reshape(100,1)
x_1_new_poly = poly_features.transform(x_1_new)
y_1_new = lin_reg.predict(x_1_new_poly)
plt.plot(x_1, y_1, 'b.')
plt.plot(x_1_new, y_1_new, "r-", linewidth = 2, label ="Predictions1") 
x_2_new = np.linspace(-2, 2, 100).reshape(100,1)
x_2_new_poly = poly_features.transform(x_2_new)
y_2_new = lin_reg.predict(x_2_new_poly)
plt.plot(x_2, y_2, 'b.')
plt.plot(x_2_new, y_2_new, "r--", linewidth = 1, label ="Predictions2") 
plt.xlabel('$x$', fontsize = 18)
plt.ylabel('$y$', rotation = 0, fontsize = 18)
plt.legend(loc = 'upper left', fontsize = 14)
plt.title('Predictions_plot')
plt.show()
```
